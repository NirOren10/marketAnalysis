```python
def clean(df):
    df['datetime'] = pd.to_datetime(df["Datetime"])
    df['price'] = df["Close"]
    df.drop(columns=["Datetime","High","Open","Close","Low"],inplace = True)
    df["moving_average"] = df["price"].rolling(window=10).mean()
    df["short_mov_avg"] = df["price"].rolling(window=10).mean()
    df["long_mov_avg"] = df["price"].rolling(window=50).mean()
    df['mov_avg_deviation'] = ((df['short_mov_avg'] - df['long_mov_avg']) / df['long_mov_avg']) * 100

    return df
```
